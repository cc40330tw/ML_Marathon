```python
import numpy as np
import matplotlib.pyplot as plt
# np.linspace(0, 100, 101)是指 0~100 劃分成 101 個刻度(含頭尾), 所也就是 0, 1, 2,...,100 這 101 個數
x_lin = np.linspace(0,100,101)
print(x_lin)

'''
# np.linspace(0, 100, 99)是指 0~100 這 101 個數當中，劃出100個數(即101除以100=1.0101...)
x_lin = np.linspace(0,100,100)
print(x_lin)

# np.linspace(0, 100, 99)是指 0~100 這 101 個數當中，劃出99個數(即101除以99=1.0204...)
x_lin = np.linspace(0,100,99)
print(x_lin)
'''
```

```python
'''Y = np.random.randn(101)
print(Y)
YY = Y * 5
print(YY)
print(x_lin + YY)
(x_lin + YY)*w + b'''

w = 3
b = 0.5
# np.random.randn()會隨機傳回"標準常態分布"的取樣值，有負有正
y = (x_lin + np.random.randn(101) * 5) * w + b 
print(y)
```

```python
plt.plot(x_lin,y,'gx',label='data points') #'gx'代表green且每個點以x表示
plt.title("Assume we have data points")

#https://blog.csdn.net/helunqu2017/article/details/78641290
plt.legend(loc='best',title='figure 1 legend') #loc='best'代表圖示那個框框最佳的擺放位置
plt.show()
```

```python
y_hat = x_lin * w + b #這邊的y_hat(即圖中紅色實線)就沒有隨機的部分了

plt.plot(x_lin, y, 'b+', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')

plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

```python
#============MAE using def ===============

def mean_absolute_error(y,yp): # y是實際值；yp是預測值
    # MAE : 將兩個陣列相減後, 取絕對值(abs), 
    #再將整個陣列加總成一個數字(sum), 最後除以y的長度(len), 
    #因此稱為"平均絕對誤差"
    mae = MAE = sum(abs(y-yp)) / len(y)
    return mae

# 呼叫上述函式, 傳回 y(藍點高度)與 y_hat(紅線高度) 的 MAE
MAE = mean_absolute_error(y,y_hat)
print("The Mean absolute error is %.3f" %(MAE))
```

```python
#============MSE <法一> using def ===============

def mean_square_error(y,yp):
    mse = MSE = sum((y-yp)**2) / len(y)
    return mse
MSE = mean_square_error(y,y_hat)
print("The Mean square error is %.3f" %(MSE))
```

```python
#============MSE <法二> using scikit-learn============

from sklearn.metrics import mean_squared_error
mean_squared_error(y,y_hat)
```

```python
#============MSE <法三> using numpy module============

import numpy as np
np.square(np.subtract(y,y_hat)).mean()
```
